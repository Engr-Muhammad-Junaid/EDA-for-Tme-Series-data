# Time Resampling
1. use resampling in datatime serieshttps:
2. //towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
 https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78

```python
#for each year we have shown the max value for each columns 
df_tesla.resample(rule='A').max()
```

```python
# A mean year frequancy(the end of the year)
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
#in each year it will take months with increase of 3 like if first months start from 1 then it will take 4 ,7,10 in each year.
df_tesla.resample(rule='QS')['High'].min().plot()
```

```python
df_tesla['High'].resample(rule='BA').max().plot(kind='barh')
```

```python
df_tesla['Low'].resample(rule='y').min().plot(kind='barh')
```

```python
df_tesla['high:10 days roiling']=df_tesla['High'].rolling(10).mean()
```

```python
df_tesla[['High','high:10 days roiling']].plot(figsize=(12,5))
```
