```python
import pandas as pd
import pandas_datareader as pdr
from datetime import datetime
```

```python
df_tesla=pdr.get_data_yahoo('TSLA')
```

```python
df_tesla
```

```python
#simple moving average.it's used for smothness of graph
df_tesla['Open'].plot()
df_tesla['Open'].rolling(10,1).mean().plot(xlim=['2020-01-01','2021-01-01'],color='red',figsize=(12,6))
```

```python
df_tesla['Open'].rolling(window=5,min_periods=2).mean().head(10)
```

```python
##commulating moving avg using expanding
df_tesla['Open'].plot()
df_tesla['Open'].expanding().mean().plot(figsize=(12,6))
```
